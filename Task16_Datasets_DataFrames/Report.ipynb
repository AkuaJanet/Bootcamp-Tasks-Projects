```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', sep = ' ')
```

Write the code needed to produce a report that provides the following information:

* Compare the average income based on ethnicity.

```python
# Group the dataset by ethnicity, call the mean method and store in a variable.
# Then use iloc to specify just the income column.
ave_income = df.groupby(['Ethnicity']).mean(numeric_only=True)
round(ave_income.iloc[:,1], 2)
```

* On average, do married or single people have a higher balance?

```python
# Separate married and single people from the married column and store in their respective variables.
# Specify only the balance column on each group and call the mean method. Use f-string to display the result. 
married = df[df['Married'] == 'Yes']
single = df[df['Married'] == 'No']

f"The average balance for married people is {round(married['Balance'].mean(), 2)} \
while the average balance for single people is {round(single['Balance'].mean(), 2)} \
so single people have a higher balance"
```

* What is the highest income in our dataset?

```python
# Specify the income column and call the max method. 
f"The highest income in the dataset is {df['Income'].max()}"
```

* What is the lowest income in our dataset?

```python
# Specify the income column and call the min method. 
f"The lowest income in the dataset is {df['Income'].min()}"
```

* How many cards do we have recorded in our dataset? (Hint: use `sum()`)

```python
# Specify the cards column and call the sum method to give the total number of cards.
f"The number of cards recorded in the dataset is {df['Cards'].sum()}"
```

* How many females do we have information for vs how many males? (Hint: use `count()`

```python
# Group the dataset by gender and call the count method.
gender_group = df.groupby('Gender').count()
gender_group
```
